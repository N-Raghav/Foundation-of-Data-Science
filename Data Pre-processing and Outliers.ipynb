# Outliers

```python
col=['Pclass','Fare']
```

```python
df= dataset[col]
```

```python
df
```

```python
grp1=df.groupby('Pclass')
```

```python
grp1.boxplot()
```

```python
def IQR(data,col):
    q1=data[col].quantile(0.25)
    q3=data[col].quantile(0.75)
    iqr=q3-q1
    return iqr,q3,q1
```

```python
iqr,q3,q1=IQR(df,'Fare')
```

```python
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
lower
```

```python
upper
```

```python
df =df[(df['Fare']>lower) & (df['Fare']<upper)]
```

```python
grp2=df.groupby('Pclass')
```

```python
grp2.boxplot()
```
