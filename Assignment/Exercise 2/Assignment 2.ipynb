```python
def IQR(data,column):
    q1=data[column].quantile(0.25)
    q3=data[column].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    return iqr,q1,q3,upper,lower
```

```python
iqr,q1,q3,upper,lower=IQR(df,"Engine HP")
df=df[(df['Engine HP']>lower) & (df['Engine HP']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"Engine Cylinders")
df=df[(df['Engine Cylinders']>lower) & (df['Engine Cylinders']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"Number of Doors")
df=df[(df['Number of Doors']>lower) & (df['Number of Doors']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"Number of Doors")
df=df[(df['Number of Doors']>lower) & (df['Number of Doors']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"highway MPG")
df=df[(df['highway MPG']>lower) & (df['highway MPG']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"city mpg")
df=df[(df['city mpg']>lower) & (df['city mpg']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"Popularity")
df=df[(df['Popularity']>lower) & (df['Popularity']<=upper)]
```

```python
iqr,q1,q3,upper,lower=IQR(df,"MSRP")
df=df[(df['MSRP']>lower) & (df['MSRP']<=upper)]
```
