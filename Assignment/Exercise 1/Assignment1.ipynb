```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("employees.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
dataset=data.dropna(axis=0,subset=['First Name','Gender','Start Date','Last Login Time','Senior Management','Team'])
```

```python
dataset.tail()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include=[int,float])
```

```python
dic={'count':dataset.count(axis=0,numeric_only=True),'max':dataset.max(axis=0,numeric_only=True),'min':dataset.min(axis=0,numeric_only=True),'mean': dataset.mean(axis=0,numeric_only=True),'std deviation': dataset.std(axis=0,numeric_only=True)}
```

```python
display=pd.DataFrame(dic)
```

```python
stores=pd.read_excel("stores.xlsx")
```

```python
stores.info()
```

```python
stores['shipment_days']=(stores['Ship Date']-stores['Order Date'])/np.timedelta64(1,'D')
```

```python
stores
```

```python
import matplotlib.pyplot as plt
```

```python
region=stores['Region'].value_counts()
print(region)
plt.bar(region.index,region) ;
```

```python
cat=stores['Category'].value_counts()
print(cat)
plt.bar(cat.index,cat) ;
```

```python
sdays=stores['shipment_days'].value_counts()
print(sdays)
plt.bar(sdays.index,sdays) ;
```

```python
sm=stores['Ship Mode'].value_counts()
print(sm)
plt.bar(sm.index,sm) ;
```
